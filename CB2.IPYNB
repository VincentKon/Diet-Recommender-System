```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

# Download required NLTK data files
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')

# Initialize the lemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

# Function to map NLTK POS tags to WordNet POS tags
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to NOUN if no match

def preprocess_text(tokens):
    # Lowercase and remove stop words
    filtered_tokens = [token.lower() for token in tokens if token.lower() not in stop_words]
    # POS tagging
    pos_tags = nltk.pos_tag(filtered_tokens)
    # Lemmatization
    lemmatized_tokens = [
        lemmatizer.lemmatize(token, get_wordnet_pos(pos)) for token, pos in pos_tags
    ]
    # Stemming
    stemmed_tokens = [stemmer.stem(token) for token in lemmatized_tokens]
    return stemmed_tokens
```

```python
import ast
def string_to_list(s):
    try:
        # Use ast.literal_eval to safely evaluate the string to a Python object
        return ast.literal_eval(s)
    except (ValueError, SyntaxError):
        # If evaluation fails, return the original string
        return s

# Apply the conversion to the 'Tags' column
tag_df['Name_x'] = tag_df['Name_x'].apply(string_to_list)
tag_df['Description_x'] = tag_df['Description_x'].apply(string_to_list)
tag_df['RecipeCategory_x'] = tag_df['RecipeCategory_x'].apply(string_to_list)
tag_df['Keywords_x'] = tag_df['Keywords_x'].apply(string_to_list)
tag_df['RecipeIngredientParts_x'] = tag_df['RecipeIngredientParts_x'].apply(string_to_list)
tag_df['RecipeInstructions_x'] = tag_df['RecipeInstructions_x'].apply(string_to_list)

# Print the DataFrame to see the results
tag_df.head()
```

```python
tag_df["Tags"] = tag_df["Name_x"] + tag_df["Description_x"] + tag_df["RecipeCategory_x"] + tag_df["Keywords_x"] + tag_df["RecipeIngredientParts_x"] + tag_df["RecipeInstructions_x"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer()
# Fit and transform the keywords column
tag_df['Proccesed_Tags'] = tag_df['Proccesed_Tags'].apply(lambda x: ' '.join(x))
tfidf_matrix = vectorizer.fit_transform(tag_df['Proccesed_Tags'])
```

```python
def get_wordnet_pos(tag):
    """Map POS tag to first character lemmatize() accepts"""
    tag_dict = {
        'J': 'a',  # Adjective
        'N': 'n',  # Noun
        'V': 'v',  # Verb
        'R': 'r'   # Adverb
    }
    return tag_dict.get(tag[0], 'n')

def preprocess_input(user_input):
    # Lowercase
    user_input = user_input.lower()
    # Remove non-alphabetic characters and keep spaces
    user_input = re.sub(r'[^a-zA-Z\s]', '', user_input)
    # Tokenize
    tokens = word_tokenize(user_input)
    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatize tokens
    pos_tags = nltk.pos_tag(tokens)
    lemmatized_tokens = [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tags]
    # Join tokens back into a single string
    tokens = [stemmer.stem(token) for token in lemmatized_tokens]
    processed_input = ' '.join(tokens)
    print(processed_input)
    return processed_input


def recommend_recipes(user_input, vectorizer_matrix):
    # Preprocess the user input
    processed_input = preprocess_input(user_input)
    # Transform the user input to match the TF-IDF matrix
    user_input_tfidf = vectorizer.transform([processed_input])
    
    # Calculate cosine similarity between user input and all recipes
    cosine_similarities = cosine_similarity(user_input_tfidf, tfidf_matrix).flatten()
    
    # Get indices of the most similar recipes (sorted by similarity score)
    similar_indices = cosine_similarities.argsort()[::-1][:3]
    
    # Get the most similar recipes
    recommendations = tag_df.iloc[similar_indices]
    # for i, r in enumerate(recommendations):
    #     print(f"{i+1}. {r[]}")
    
    return recommendations

# Example user input
user_input = "healthy summer dessert"

# Get recommendations
recommended_recipes = recommend_recipes(user_input, tfidf_matrix)
recommended_recipes
```
