```python
import pandas as pd
import numpy as np
import pickle
```

```python
recipe_df = pd.read_csv("data/Foodcom/recipes.csv")
rating_df = pd.read_csv("data/Foodcom/reviews.csv")
```

```python
num_rating_df = rating_df.groupby("RecipeId").count()["Rating"].reset_index()
num_rating_df.rename(columns={"Rating": "Num_Ratings"}, inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_df.groupby("RecipeId").mean()["Rating"].reset_index()
avg_rating_df.rename(columns={"Rating": "Avg_Ratings"}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="RecipeId")
popular_df
```

```python
popular_df = popular_df[popular_df["Num_Ratings"]>250].sort_values("Avg_Ratings", ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(recipe_df, on="RecipeId")
popular_df.info()
```

```python
pickle.dump(popular_df, open("model/popular.pickle", "wb"))
```
