```python
def extract_first_url(url_string):
    # Check if url_string is a string
    if isinstance(url_string, str):
        # Remove 'c(' at the beginning and ')' at the end, then split by ', '
        urls = url_string.strip('c()').replace('"', '').split(', ')
        # Return the first URL
        return urls[0]
    else:
        # If url_string is not a string, return it as is (could be NaN)
        return url_string

# Apply the function to the 'Images' column for all rows
df['Images'] = df['Images'].apply(extract_first_url)

# Print the modified DataFrame
a = (df["Images"][0])
print(a)
```
